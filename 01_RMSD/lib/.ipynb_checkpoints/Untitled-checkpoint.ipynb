```python
# load dependencies
import pandas as pd
import pyreadr
import matplotlib.pylab as plt
import numpy as np
from sklearn.mixture import GaussianMixture
from scipy.stats import norm

%matplotlib inline
```

```python
# Read in Rdata file
result = pyreadr.read_r('../data/Simons_exp_1/rmsd_kmer_8.Rdata')
df = result[list(result.keys())[0]]
```

```python
# Mutate to fit original meaning
clean_df = pd.DataFrame({
    'x': df.index.values - 300.5,
    'RMSD': df[None].values
})
```

```python
clean_df.head()
```

```python
# Plot clean_df
plt.rcParams["figure.figsize"] = (10,7)
plt.plot(clean_df['x'], clean_df['RMSD'])
```

```python
# Fit gaussian mixture model to data
X = clean_df['RMSD'].values
gm = GaussianMixture(n_components=3, random_state=0).fit(X.reshape(-1, 1))
```

```python
# Represent the gaussian mixture with plots
# plt.hist(X, bins=50, density=True)
xs = np.arange(min(X)-1, max(X)+1, 0.05)
for mean, sd, weight in zip(gm.means_, np.sqrt(gm.covariances_), gm.weights_):
    plt.plot(xs, norm.pdf(xs, mean, sd).reshape(-1) * weight)
plt.ylim(0, 1e-6)
plt.xlim(-5, 5)
plt.show()
```
